## Introduction

Consider a sequence of independent random variables $(y_k)_k$ with probability density $p_\theta(y)$ where $\theta $ is scalar parameters. We assume that paramaters before change i.e. $\theta$ is known. Log likelihood ratio is defined as 
$$s(y) = \ln\frac{p_{\theta_1}(y)}{p_{\theta_0}(y)}$$, 
Key statistical property of this ratio is:
$$ E_{\theta_0}(s) < 0 \text{  and  } E_{\theta_1}(s) > 0 $$

*Change in the parameter $\theta$ is reflected as a change in the sign of the mean value of the log-likelihood ratio.*
Kullback information is $K(\theta_1, \theta_0) = E_{\theta_1}(s)$, hence Kullback information can be also used as detectibility. 


## Elementary Algorithms
### Limit Checking Detectors and Shewhart Control Charts
#### Continuous Inspection in quality control
* Samples with fixed size N are taken.
* At the end of each sample a decision rule is computed to test below hypothesis about $\theta$ :
$$ H_0 \text{ : } \theta = \theta_0 $$
$$ H_1 \text{ : } \theta = \theta_1 $$

* As long as decision is taken in favour of *H0*, the sample coninues.

*Notations*

$ s_i $   : log-likelihood ratio for the observations $y_i$ *sufficient statistics*
$$ s_i = \ln \frac{p_{\theta_1}(y)}{p_{\theta_1}(y)} $$
$ S_j^k $ : log-likelihood ratio for the observations from $y_j$ to $y_k$.
$$ S_j^k = \sum_{i=j}^k s_i $$

$ S_1^N $ : decision function

$h $      : threshold

$ d $     : optimal decision rule
\begin{equation}
    d=
    \begin{cases}
      0, & \text{if} S_1^N < h \text{; } H_0 \text{ is chosen} \\
      1, & \text{if} S_1^N \ge h \text{; } H_1 \text{ is chosen}
    \end{cases}
\end{equation}
$ d_K $   : deciosion rule for the sample K( of size N)

$t_a $    : alarm time, 
$$ t_a = N.min\{K:d_K = 1\} $$

#### Example for Gaussian(Change in mean)
Consider particular case where distribution is Gaussian and mean is $\mu$ and constant variance $\sigma^2$. In this case changing parameter $\theta $ is $\mu $. 
$$ p_\theta(y) = \frac{1}{\sigma\sqrt{2\pi}} e ^ {- \frac{(y - \mu)^2}{2\sigma^2}}$$
and sufficient statistics $s_i$ is
$$ s_i = \frac{\mu_1 - \mu_0}{\sigma^2} \left (y_i - \frac {\mu_0 + \mu_1}{2} \right) $$
which can be rewritten as 
$$ s_i = \frac{b}{\sigma} \left (y_i - \frac {\mu_0 + \mu_1}{2} \right) \\ 
= \frac{b}{\sigma} \left (y_i - \mu_0 + \frac {\nu}{2} \right) $$
where, 
$$ \nu = \mu_1 - \mu_0 \text{ is the change in magnitude}\\ 
b = \frac{\mu_1 - \mu_0}{\sigma} \text{is signal to noise ratio} $$
therefore the decision function is 
$$ S_1^N = \frac{b}{\sigma}\sum_{i=1}^N\left(y_i-\mu_0 - \frac{\nu}{2}\right) $$

*The stopping rule for the change detection algorithm is, with the decision rule defined by*  
\begin{equation}
    d=
    \begin{cases}
      0, & \text{if} S_1^N(K) < h  \\
      1, & \text{if} S_1^N(K) \ge h
    \end{cases}
\end{equation}
